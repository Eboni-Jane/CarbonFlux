```python
at_bins = np.linspace(280,310)
at_binmids = at_bins[1:]-np.diff(at_bins)[0]/2
ev_binmeans = np.empty((len(at_bins)-1))
ev_bin25 = np.empty((len(at_bins)-1))
ev_bin75 = np.empty((len(at_bins)-1))
for i in range(len(at_bins)-1):
    ev_binmeans[i] = np.mean(ev[(at>at_bins[i])&(at<=at_bins[i+1])])
    ev_bin25[i] = np.percentile(ev[(at>at_bins[i])&(at<=at_bins[i+1])],25)
    ev_bin75[i] = np.percentile(ev[(at>at_bins[i])&(at<=at_bins[i+1])],75)
```

```python
plt.scatter(at,ev,s=3)
plt.plot(at_binmids,ev_binmeans,'r')
plt.fill_between(at_binmids,ev_bin25,ev_bin75,color='r',alpha=0.5,lw=0)
```
