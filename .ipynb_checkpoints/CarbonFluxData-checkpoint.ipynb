# Packages

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import scipy.stats as stats
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = 'Arial' # Choose a nice font here
plt.rcParams['font.size'] = 7 # standard font size
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.it'] = 'Arial'
plt.rcParams['figure.facecolor'] = 'white' # make sure background isn't transparent
plt.rcParams['figure.dpi'] = 150
```

# Functions and constants

```python
def Temperatureflux(t,h):
    return t/h
```

# Read data

```python
data = pd.read_csv('Variables.csv')
H   = data['Hours']
T   = data['Temp']
ET  = data['ET']
CO2  = data['CO2']
CH4  = data['CH4']
```

```python
data_Hourly = pd.read_csv('Hourly Variables.csv')
h   = data_Hourly['Hours']
t   = data_Hourly['Temp']
et  = data_Hourly['ET']
co2  = data_Hourly['CO2']
ch4  = data_Hourly['CH4']
```

```python
data_raw = pd.read_csv('raw data.csv')
d   = data_raw['DOY']
at   = data_raw['air_temperature']
ev  = data_raw['ET']
cf  = data_raw['co2_flux']
```

```python
data_ch4_raw = pd.read_csv('CH4 raw.csv')
D   = data_ch4_raw['DOY']
AIRT   = data_ch4_raw['air_temperature']
MF  = data_ch4_raw['ch4_flux']
```

# Plots

```python
fig = plt.gcf()

plt.subplot(221)
plt.plot(h, t)
plt.xlabel('Day (Hrs)')
plt.ylabel('Temperature (K)')

plt.subplot(222)
plt.plot(h, et)
plt.xlabel('Day (Hrs)')
plt.ylabel('Evapotranspiration ()')

plt.subplot(223)
plt.plot(h, co2)
plt.xlabel('Day (Hrs)')
plt.ylabel('CO2 ()')

plt.subplot(224)
plt.plot(h, ch4)
plt.xlabel('Day (Hrs)')
plt.ylabel('CH4 ()')

plt.subplots_adjust(hspace=0.3,wspace=0.2)
fig.set_size_inches(10,5)
```

```python
fig = plt.gcf()

plt.subplot(221)
plt.plot(d, at)
plt.xlabel('Year (Days)')
plt.ylabel('Temperature (K)')

plt.subplot(222)
plt.plot(d, ev)
plt.xlabel('Year (Days)')
plt.ylabel('Evapotranspiration ()')

plt.subplot(223)
plt.plot(d, cf)
plt.xlabel('Year (Days)')
plt.ylabel('CO2 ()')

plt.subplot(224)
plt.plot(D, MF)
plt.xlabel('Year (Days)')
plt.ylabel('CH4 ()')

plt.subplots_adjust(hspace=0.3,wspace=0.2)
fig.set_size_inches(10,5)
```

```python
at_bins = np.linspace(280,310)
at_binmids = at_bins[1:]-np.diff(at_bins)[0]/2
ev_binmeans = np.empty((len(at_bins)-1))
ev_bin25 = np.empty((len(at_bins)-1))
ev_bin75 = np.empty((len(at_bins)-1))
for i in range(len(at_bins)-1):
    ev_binmeans[i] = np.mean(ev[(at>at_bins[i])&(at<=at_bins[i+1])])
    ev_bin25[i] = np.percentile(ev[(at>at_bins[i])&(at<=at_bins[i+1])],25)
    ev_bin75[i] = np.percentile(ev[(at>at_bins[i])&(at<=at_bins[i+1])],75)
```

```python
plt.scatter(at,ev,s=3)
plt.plot(at_binmids,ev_binmeans,'r')
plt.fill_between(at_binmids,ev_bin25,ev_bin75,color='r',alpha=0.5,lw=0)
```
