```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import scipy.stats as stats
import seaborn as sns
from tabulate import tabulate
import plotly.graph_objects as go
```

```python
data_m = pd.read_csv('CH4_Mean.csv')
t  = data_m['time']
sm  = data_m['Summer_ch4_flux']
spm = data_m['Spring_ch4_flux']
am  = data_m['Autumn_ch4_flux']
etaticks = [0, 10, 20, 30, 39, 47]
```

```python
data_m = pd.read_csv('CH4_Winter_Mean.csv')
T  = data_m['time']
wm  = data_m['ch4_flux']
```

```python
q = sm < 0
sm[q]
len(sm[q])
sm[q].sum()
```

```python
w = sm > 0
sm[w]
len(sm[w])
sm[w].sum()
```

```python

plt.plot(t, sm, color = 'k', alpha = 0.6)
plt.xlabel('Time (Hrs)')
plt.ylabel('CH4 (${\mu}$mol$^{-2}$s$^{-1}$)')
plt.fill_between(t,sm, where = sm > 0, alpha=0.30, color='red', interpolate=True, label = 'CH4 flux out: 0.43')
plt.fill_between(t, sm, where = sm < 0, alpha=0.30, color='c', interpolate=True, label = 'CH4 flux in: -0.24')
plt.xlim(etaticks[0],etaticks[-1])
plt.xticks(etaticks)
plt.legend(frameon=False)
```
